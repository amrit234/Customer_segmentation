K-means clustering is a machine learning algorithm used for partitioning data into distinct groups or clusters based on similarities in features. In customer segmentation projects, K-means is utilized to group customers with similar characteristics, behaviors, or preferences. It helps businesses understand their customer base, tailor marketing strategies, and personalize services to meet specific needs effectively.

```python
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.metrics import silhouette_score
```

```python
X = df2.drop('customer_id', axis=1)
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)
```

```python
kmeans = KMeans(n_clusters=3)
train_labels = kmeans.fit_predict(X_train)
```

```python
test_labels = kmeans.predict(X_test)
```

```python
train_silhouette_score = silhouette_score(X_train, train_labels)
test_silhouette_score = silhouette_score(X_test, test_labels)

print(f"Training Silhouette Score: {train_silhouette_score}")
print(f"Testing Silhouette Score: {test_silhouette_score}")
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
from sklearn.model_selection import train_test_split
```

```python
X = df2.drop('products_purchased', axis=1)
```

```python
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)
n_clusters = 3
hierarchical = AgglomerativeClustering(n_clusters=n_clusters)
train_labels = hierarchical.fit_predict(X_train)
```

```python
test_labels = hierarchical.fit_predict(X_test)
```

```python
train_silhouette_score = silhouette_score(X_train, train_labels)
test_silhouette_score = silhouette_score(X_test, test_labels)

print(f"Training Silhouette Score: {train_silhouette_score}")
print(f"Testing Silhouette Score: {test_silhouette_score}")
```

```python
from sklearn.mixture import GaussianMixture
```

```python
X = df2.drop('complains', axis=1)
```

```python
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)
```

```python
n_components = 3
gmm = GaussianMixture(n_components=n_components)
train_labels = gmm.fit_predict(X_train)
```

```python
test_labels = gmm.predict(X_test)
```

```python
train_silhouette_score = silhouette_score(X_train, train_labels)
test_silhouette_score = silhouette_score(X_test, test_labels)

print(f"Training Silhouette Score: {train_silhouette_score}")
print(f"Testing Silhouette Score: {test_silhouette_score}")
```
